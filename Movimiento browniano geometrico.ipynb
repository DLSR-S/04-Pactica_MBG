```python
import pandas as pd 
import numpy as np
import math
import random
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
def xt(miu = .027, sigma=.37, dt=1/48, spot=2.95):
  uniforme = []
  z = []
  browniano = []
  dxt = []
  dt = dt
  miu = miu
  sigma = sigma
  spot = spot
  trayectoria = [spot]

  for x in range(1,48):
    uniforme_num = random.random()
    uniforme.append(uniforme_num)
    normal = norm.ppf(uniforme_num)
    z.append(normal)
    dwt = dt**.5*normal
    browniano.append(dwt)
    dxt_num = miu*dt*trayectoria[x-1]+sigma*trayectoria[x-1]*dwt
    dxt.append(dxt_num)
    trayectoria_num = dxt_num + trayectoria[x-1]
    trayectoria.append(trayectoria_num)
  
  return trayectoria
```

```python
def creando_diccionario():
  dic_com = {}
  for x in range(1,450):
    arr= xt()
    dic_com[x] = arr
  
  return dic_com
```

```python
dicc = creando_diccionario()
df1 = pd.DataFrame(dicc)
```
